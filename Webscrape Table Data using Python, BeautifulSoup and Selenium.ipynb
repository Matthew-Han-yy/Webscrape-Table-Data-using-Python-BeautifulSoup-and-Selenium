```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
from bs4 import BeautifulSoup
import pandas as pd

# ceate 3 strings
loginuserStr = 'test'
loginpwdStr = 'test'
loginvcodeStr = '888888'

# the website address
url = 'http://149.28.137.102:89'

# chromedriver path
chromedriver = "/Users/matthewhan/Desktop/Intro_Machinelearning/chromedriver/chromedriver"
os.environ["webdriver.chrome.driver"] = chromedriver

# instruct browser to open up Google Chrome and navigate to the login page
browser = webdriver.Chrome(chromedriver)
time.sleep(1)
browser.get(url)
```

Chrome is now being controlled by automated test software Selenium.

```python
# After successfully gotten the browser to open at the site, we need to find the input fields on the webpage, into which we enter the 3 strings above.
username_input_field = browser.find_element_by_id('username')
username_input_field.send_keys(loginuserStr)

pwd_input_field = browser.find_element_by_id('pwd')
pwd_input_field.send_keys(loginpwdStr)

code_input_field = browser.find_element_by_id('code')
code_input_field.send_keys(loginvcodeStr)

loginButton = browser.find_element_by_id('loginin')
loginButton.click()
```

The send_keys() and click() commands simulates keypresses and mouse click in the input field respectively.

Now we have successfully entered into the webpage where we have the information in the table to be scrapped.

Now Selenium hands the page source to Beautiful Soup.

```python
# create beautifulsoup object and prettify function in BeautifulSoup enable us to view how the tags are nested in the document.
soup = BeautifulSoup(browser.page_source, 'lxml')
print(soup.prettify())
```

```python
# locate the headers and store them into a list
# this will form the column names for our dataframe later. 
headers = []
for th in soup.find_all('th'):
    headers.append(th.get_text())
headers
```

```python
# locate the data in <td>, we need data from position [2] to [31]
td = soup.find_all('td')
data = []
for d in td[2:32]:
    data.append(d.get_text())
print(data)
```

```python
# there are 3 rows of data so separate them into 3 lists
data_table = [data[:10], data[10:20], data[20:]]

# convert data_table into dataframe
df = pd.DataFrame(data_table, columns=headers)
df
```
